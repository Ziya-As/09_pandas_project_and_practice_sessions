```python
# clean up data
```

```python
# Only races 50km or 50mile
df[df["Event distance/length"].isin(["50km", "50mi"])]
```

```python
# Only races 50km or 50mile, 2020
df[ (df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020)]
```

```python
# Only races in USA
df[df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA"]
```

```python
# Only USA races, 50km or 50mile, 2020
df[ (df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA")]
```

```python
df2 = df[ (df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA")]
```

```python
df2.loc[:, "athlete_age"] = 2020 - df2["Athlete year of birth"]
```

```python
# drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2 = df2.drop(columns=["Athlete club", "Athlete country", "Athlete year of birth", "Athlete age category"])
```

```python
# check for duplicates
df2[df2.duplicated()]
```

```python
df2["athlete_age"] = df2["athlete_age"].astype(int)
```

```python
# questions to find out from the data
```

```python
# Difference in speed for the 50km, 50mi male to female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age groups are the best in the 50m race (20 + races min)
df3\
  .query('race_length == "50mi"')\
    .groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count'])\
        .sort_values('mean', ascending=False)\
          .query('count > 19').head(15)
```

```python
# What age groups are not the best in the 50m race (20 + races min) (show 20)
df3\
  .query('race_length == "50mi"')\
    .groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count'])\
    .sort_values('mean', ascending=True)\
    .query('count > 10').head(20)
```

```python
# Seasons for the data - Are runners slower in summer than winter?
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(
  lambda x: 'Winter' if x > 11 
    else 'Fall' if x > 8 
    else 'Summer' if x > 5 
    else 'Spring' if x > 2 
    else 'Winter'
  )
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
# 50 miler only
df3.query('race_length == "50mi"')\
  .groupby('race_season')['athlete_average_speed'].agg(['mean', 'count'])\
  .sort_values('mean', ascending=False)
```
